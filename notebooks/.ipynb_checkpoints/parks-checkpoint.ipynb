```python
import sys
sys.path.append('..')
import json
import folium
import scrapper.scrap_parks as sc
```

```python
df = sc.scrap_park_data()
df.tail(5)
```

```python
park = 'yell'
geojson = df[df['parkunit']==park]['boundaries'].values[0]
lat = df[df['parkunit']==park]['latitude'].values[0]
long = df[df['parkunit']==park]['longitude'].values[0]
```

```python
data = json.loads(geojson)
```

```python
m = folium.Map(location=[lat, long])

folium.GeoJson(
    data,
    name='geojson'
).add_to(m)

folium.LayerControl().add_to(m)

m
```
